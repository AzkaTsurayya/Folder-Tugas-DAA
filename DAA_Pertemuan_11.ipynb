```python
class graph:
  def __init__(self, gdict=None):
    if gdict is None:
      gdict = {}
    self.gdict = gdict

  def edges(self):
    return self.findedges()

# Add the new edge
  def AddEdge(self, edge):
    edge = set(edge)
    (vrtx1, vrtx2) = tuple(edge)
    if vrtx1 in self.gdict:
      self.gdict[vrtx1].append(vrtx2)
    else:
      self.gdict[vrtx1] = [vrtx2]

# List the edge names
  def findedges(self):
    edgename = []
    for vrtx in self.gdict:
      for nxtvrtx in self.gdict[vrtx]:
        if {nxtvrtx, vrtx} not in edgename:
          edgename.append({vrtx, nxtvrtx})
    return edgename

# Create the dictionary with  graph elements
graph_elements = { "a" : ["b","c"],
                "b" : ["a", "d"],
                "c" : ["a", "d"],
                "d" : ["e"],
                "e" : ["d"]
                }

g = graph(graph_elements)
g.AddEdge({'a','e'})
g.AddEdge({'a','c'})
print(g.edges())
```

```python
'''
1. Minimum Spanning Tree (MST) dengan Prim–Dijkstra & Kruskal

A. MST dengan Algoritma Prim–Dijkstra

Prim memilih edge dengan bobot terkecil yang terhubung ke node yang sudah masuk MST.

Langkah-langkah manual Prim

Pilih node awal (biasanya A).

Dari node A, pilih edge paling kecil yang keluar dari A.

Masukkan node tujuan ke MST.

Dari semua node yang sudah masuk MST, pilih edge paling kecil yang menuju node yang belum ada di MST.

Ulangi sampai semua node masuk.

Contoh awal:

Dari A, edge bobot terkecil = A–D (10)

Node dalam MST = {A, D}

Edge keluar yang mungkin: D–E (6), D–G (15), D–C (16), A–C (20)

Pilih yang terkecil = D–E (6)

MST = A–D (10), D–E (6)

Lanjutkan aturan yang sama sampai selesai.

Nanti akan kamu dapat MST total sekitar biaya minimum ± 66–70 (tergantung edge mana yang tepat, nanti aku bantu hitungkan kalau kamu mau).

B. MST dengan Algoritma Kruskal

Kruskal memilih edge terkecil dari seluruh grafik, urutkan dari yang terkecil.

Langkah:

Urutkan semua edge dari bobot terkecil → terbesar

Ambil edge terkecil tanpa membentuk siklus

Ulangi hingga semua node tersambung

Contoh urutan awal:

E–H = 4

E–F = 5

D–E = 6

F–H = 7

H–I = 8

C–F = 11

G–H = 12

B–C = 14
... dan seterusnya.

Setiap kali ambil edge, cek apakah menyebabkan siklus.
Jika ya → lewati, jika tidak → masukkan ke MST.
'''

# Graph pertama untuk MST
graph_mst = {
    'A': {'D': 10, 'C': 20},
    'B': {'C': 24, 'F': 17},
    'C': {'A': 20, 'B': 24, 'D': 16, 'E': 22, 'F': 18},
    'D': {'A': 10, 'C': 16, 'E': 6, 'G': 15},
    'E': {'D': 6, 'C': 22, 'F': 5, 'H': 4},
    'F': {'C': 18, 'B': 17, 'E': 5, 'H': 7},
    'G': {'D': 15, 'H': 12},
    'H': {'G': 12, 'E': 4, 'F': 7, 'I': 8},
    'I': {'H': 8}
}

# Graph kedua untuk Greedy Shortest Path
graph_greedy = {
    'A': {'B': 12, 'D': 14},
    'B': {'A': 12, 'C': 5, 'E': 3, 'D': 8},
    'C': {'B': 5, 'E': 9, 'F': 6},
    'D': {'A': 14, 'B': 8, 'E': 15},
    'E': {'B': 3, 'C': 9, 'D': 15, 'F': 8},
    'F': {'C': 6, 'E': 8}
}
```

```python
'''
2. Shortest Path A ke F menggunakan Metode Greedy

Greedy memilih langkah terbaik lokal (pilih edge terkecil di setiap langkah).

Cara:

Mulai dari A

Lihat semua edge dari A → pilih bobot terkecil

Pindah ke node itu

Ulangi sampai tiba di F

Dari grafik:

Dari A, edge:

A–B = 12

A–D = 14

Greedy memilih A → B (12)

Dari B, edge terkecil:

B–E = 3

B–C = 5

B–D = 8

Greedy pilih B → E (3)

Dari E, edge:

E–C = 9

E–F = 15

Greedy pilih E → C (9)

Dari C, edge:

C–F = 6

C–E = 9

C–B = 5

Greedy pilih C → F (6)

Jadi jalur Greedy:

A → B → E → C → F
Total bobot = 12 + 3 + 9 + 6 = 30
'''

import heapq

def prim(graph, start):
    visited = set()
    edges = []
    mst = []
    total_cost = 0

    visited.add(start)

    # Masukkan edge dari node start
    for dest, weight in graph[start].items():
        heapq.heappush(edges, (weight, start, dest))

    while edges:
        weight, u, v = heapq.heappop(edges)
        if v not in visited:
            visited.add(v)
            mst.append((u, v, weight))
            total_cost += weight

            for next_node, w in graph[v].items():
                if next_node not in visited:
                    heapq.heappush(edges, (w, v, next_node))

    return mst, total_cost

mst_prim, cost_prim = prim(graph_mst, 'A')
print("MST Prim:", mst_prim)
print("Total Cost:", cost_prim)

#=========== MST Kruskal ===========

# Struktur Union-Find
class UnionFind:
    def __init__(self, nodes):
        self.parent = {node: node for node in nodes}

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]

    def union(self, x, y):
        px = self.find(x)
        py = self.find(y)
        if px != py:
            self.parent[py] = px
            return True
        return False

def kruskal(graph):
    edges = []
    for u in graph:
        for v, w in graph[u].items():
            if (v, u, w) not in edges:  # Hindari duplikasi AB dan BA
                edges.append((u, v, w))

    edges.sort(key=lambda x: x[2])   # sort berdasarkan bobot

    uf = UnionFind(graph.keys())
    mst = []
    total_cost = 0

    for u, v, w in edges:
        if uf.union(u, v):
            mst.append((u, v, w))
            total_cost += w

    return mst, total_cost

mst_kruskal, cost_kruskal = kruskal(graph_mst)
print("MST Kruskal:", mst_kruskal)
print("Total Cost:", cost_kruskal)
```

```python
'''
3. Mengapa Metode Greedy Kadang Gagal Memberikan Solusi Terbaik?

Karena Greedy hanya melihat pilihan terbaik saat ini (lokal optimum) tanpa memikirkan konsekuensi langkah berikutnya.

Akibatnya, jalur yang tampak murah di awal bisa mengakibatkan total perjalanan lebih mahal dibanding jalur lain yang awalnya terlihat lebih mahal.

Contohnya:

Di grafik, Greedy memilih A–B karena 12 < 14

Tetapi jalur A–D–C–F mungkin lebih murah secara total

Greedy “terjebak” pada keputusan awal yang tampak terbaik.
'''

def greedy_shortest_path(graph, start, goal):
    path = [start]
    current = start
    total_cost = 0

    while current != goal:
        neighbors = graph[current]

        # Ambil edge dengan bobot terkecil
        next_node = min(neighbors, key=lambda x: neighbors[x])
        cost = neighbors[next_node]

        total_cost += cost
        path.append(next_node)
        current = next_node

    return path, total_cost

path, cost = greedy_shortest_path(graph_greedy, 'A', 'F')
print("Greedy Path:", path)
print("Total Cost:", cost)
```
