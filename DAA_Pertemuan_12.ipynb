# Latihan 1 Ukur Centrality

```python
'''
1. Degree Centrality
Degree = jumlah koneksi langsung.

| Node | Degree             |
| ---- | ------------------ |
| 1    | 2 (dengan 5, 2)    |
| 2    | 3 (dengan 1, 3, 4) |
| 3    | 3 (dengan 6, 2, 4) |
| 4    | 2 (dengan 3, 2)    |
| 5    | 2 (dengan 1, 6)    |
| 6    | 2 (dengan 5, 3)    |

2. Betweenness Centrality
Betweenness = seberapa sering node berada pada jalur terpendek antara dua node lain.

Jalur melibatkan node 3 sangat sering:
  - 6 → 2 = 6–3–2
  - 6 → 4 = 6–3–4
  - 5 → 2 = 5–6–3–2
  - 5 → 4 = 5–6–3–4
  - 1 → 3 = 1–2–3
  - 1 → 4 = 1–2–4

Node 2 juga sering muncul sebagai perantara:
  - 1 → 3 = 1–2–3
  - 1 → 4 = 1–2–4
  - 5 → 3 = 5–6–3 (tidak lewat 2)
  - 5 → 4 = 5–6–3–4 (tidak lewat 2)
  - 6 → 4 = 6–3–4 (tidak lewat 2)

Node paling penting sebagai penghubung: 3

3. Closeness Centrality
Closeness = seberapa dekat sebuah node ke semua node lain
(semakin kecil total jarak, semakin tinggi closeness).

Node 3
  - ke 6 = 1
  - ke 4 = 1
  - ke 2 = 1
  - ke 5 = 2
  - ke 1 = 2
Total = 7

Node 2
  - ke 1 = 1
  - ke 4 = 1
  - ke 3 = 1
  - ke 6 = 2
  - ke 5 = 3
Total = 8

Closeness tertinggi: Node 3

4. Eigenvector Centrality
Eigenvector tinggi = terhubung ke node yang juga penting.

| Ranking | Node       |
| ------- | ---------- |
| **1**   | **Node 3** |
| **2**   | Node 2     |
| 3       | Node 4     |
| 4       | Node 6     |
| 5       | Node 1     |
| 6       | Node 5     |
'''
```

```python
import networkx as nx

G = nx.Graph()
edges = [
    (5, 1),
    (5, 6),
    (6, 3),
    (3, 2),
    (3, 4),
    (2, 4),
    (1, 2)
]

G.add_edges_from(edges)

# 1. Degree Centrality
degree_centrality = nx.degree_centrality(G)

# 2. Betweenness Centrality
betweenness_centrality = nx.betweenness_centrality(G, normalized=True)

# 3. Closeness Centrality
closeness_centrality = nx.closeness_centrality(G)

# 4. Eigenvector Centrality
eigenvector_centrality = nx.eigenvector_centrality(G)

print("=== DEGREE CENTRALITY ===")
for n, v in degree_centrality.items():
    print(f"Node {n} : {v:.4f}")

print("\n=== BETWEENNESS CENTRALITY ===")
for n, v in betweenness_centrality.items():
    print(f"Node {n} : {v:.4f}")

print("\n=== CLOSENESS CENTRALITY ===")
for n, v in closeness_centrality.items():
    print(f"Node {n} : {v:.4f}")

print("\n=== EIGENVECTOR CENTRALITY ===")
for n, v in eigenvector_centrality.items():
    print(f"Node {n} : {v:.4f}")
```

# Latihan 2 Network

```python
import networkx as nx
import pandas as pd

G = nx.Graph()

nodes = ["Mike","Bob","Emma","John","Shane","Leah","Jill","Liz","Allen","Lisa"]
G.add_nodes_from(nodes)

edges = [
    ("Mike","Bob"), ("Mike","Emma"), ("Mike","Jill"),
    ("Bob","Emma"), ("Bob","John"), ("Bob","Jill"),
    ("Emma","Shane"), ("Emma","Jill"),
    ("John","Shane"), ("John","Jill"), ("John","Leah"),
    ("Shane","Jill"), ("Shane","Leah"),
    ("Leah","Jill"),
    ("Emma","Liz"),
    ("Liz","Allen"),
    ("Allen","Lisa")
]

G.add_edges_from(edges)

eig = nx.eigenvector_centrality(G)

df = pd.DataFrame.from_dict(eig, orient='index', columns=['Eigenvector'])

df_sorted = df.sort_values(by="Eigenvector", ascending=False)

print(df_sorted)
```
