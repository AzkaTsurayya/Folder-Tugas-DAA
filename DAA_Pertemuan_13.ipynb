# Implementasi Algoritma Bradth-first Search

```python
graph = {'Amin' : ['Wasim', 'Nick', 'Mike'],
         'Wasim' : ['Imran', 'Amin'],
         'Imran' : ['Wasim', 'Faras'],
         'Faras' : ['Imran'],
         'Mike' : ['Amin'],
         'Nick' : ['Amin']
         }
```

```python
def bfs(graph, start):
  visited = []
  queue = []
  queue = [start]
  while queue:
    node = queue.pop(0)
    if node not in visited:
      visited.append(node)
      neighbours = graph[node]
      for neighbour in neighbours:
        queue.append(neighbour)
  return visited
```

# Amin

```python
# O Amin ---------------
# |            |       |
# O Wasim   O Nick  O Mike
# |
# O Imran
# |
# O Faras

bfs(graph, 'Amin')
```

# Latihan 2

```python
graph = {
    'Rektor': ['Warek 1', 'Warek 2'],
    'Warek 1': ['Kaprodi 1'],
    'Warek 2': ['Kaprodi 2', 'Kaprodi 3'],
    'Kaprodi 1': ['Dosen A', 'Dosen B', 'Dosen C'],
    'Kaprodi 2': ['Dosen D', 'Dosen E'],
    'Kaprodi 3': ['Dosen F', 'Dosen G'],

    'Dosen A': set(),
    'Dosen B': set(),
    'Dosen C': set(),
    'Dosen D': set(),
    'Dosen E': set(),
    'Dosen F': set(),
    'Dosen G': set()
}
```

```python
def dfs(graph, node, visited=None):
    if visited is None:
        visited = []

    visited.append(node)

    for neighbour in sorted(graph[node]):  # urutkan supaya rapi
        if neighbour not in visited:
            dfs(graph, neighbour, visited)

    return visited

print(dfs(graph, "Rektor"))
```
