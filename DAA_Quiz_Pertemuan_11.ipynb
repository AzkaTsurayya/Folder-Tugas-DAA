```python
# Soal 1 - Hitung Jumlah Angka Genap
def hitung_genap(arr):
    return sum(1 for x in arr if x % 2 == 0)

'''
Pendekatan/Algoritma:
- Iterasi setiap elemen dalam array.
- Periksa apakah elemen habis dibagi 2(x % 2 == 0).
- Hitung jumlah elemn yang memenuhi kondisi.

Edge cases:
- Array kosong → hasil 0
- Semua bilangan ganjil → hasil 0
- Bilangan negatif tetap bisa dicek dengan % 2
'''

print(hitung_genap([1, 2, 3, 4, 7, 10]))
```

```python
# Soal 2 - Cari Nilai Terbesar
def cari_max(arr):
    if not arr:  # jika array kosong
        return None
    return max(arr)

'''
Langkah-langkah:
- Jika array kosong, kembalikan None atau pesan khusus.
- Gunakan fungsi built-in max() atau lakukan iterasi manual untuk mencari nilai terbesar

Edge cases:
- Array kosong → harus ditangani
- Semua elemen sama → tetap mengembalikan nilai tersebut
'''

print(cari_max([5, 2, 9, 1]))
```

```python
# Soal 3 - Balik String
def balik_string(s):
    return s[::-1]

'''
Logika:
- Gunakan slicing [;;-1] untuk membalik urutan karakter
- Alternatif: iterasi dari belakang ke depan
'''

print(balik_string("halo"))
```

```python
# Soal 4 - Jumlahkan Semua Elemen
def jumlahkan(arr):
    return sum(arr)

'''
Langkah-langkah:
- Iterasi setiap elemen
- Tambahkan ke total akumulasi
- Kembalikan hasil akhir
'''

print(jumlahkan([3, 1, 2]))
```

```python
# Soal 5 - Cek Palindrome
def cek_palindrome(s):
    return s == s[::-1]

'''
Cara menentukan:
- Bandingkan string dengan versi terbaliknya.
- Jika sama → palindrome.
'''

print(cek_palindrome("level"))  # True
print(cek_palindrome("hello"))  # False
```

```python
# Soal 6 - Hitung Jumlah Huruf Vokal
def hitung_vokal(s):
    vokal = "aiueoAIUEO"
    return sum(1 for ch in s if ch in vokal)

'''
Cara memeriksa:
- Iterasi setiap karakter.
- Cek apakah karakter termasuk dalam daftar vokal.
- Tambahkan ke hitungan
'''

print(hitung_vokal("informatika"))
```

```python
# Soal 7 - Temukan Elemen yang Dicari
def cari_target(arr, target):
    return target in arr

'''
Metode pencarian:
- Gunakan operator in (linear search).
- Alternatif: iterasi manual dan bandingkan setiap elemen.
'''

print(cari_target([4, 7, 1, 9], 7))
```
